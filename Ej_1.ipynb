```python
import boto3

# ==============================
# CONFIGURACIÓN
# ==============================
bucket_name = "practica-milton-aws-123456"  # 
file_name = "archivo_prueba.txt"
object_name = "archivo_prueba.txt"
region = "us-west-2"

# ==============================
# CREAR CLIENTE S3
# ==============================
s3 = boto3.client('s3', region_name=region)

print("Cliente S3 inicializado")

# ==============================
# CREAR BUCKET (si no existe)
# ==============================
try:
    s3.create_bucket(
        Bucket=bucket_name,
        CreateBucketConfiguration={
            'LocationConstraint': region
        }
    )
    print("Bucket creado correctamente")
except Exception as e:
    print("Bucket puede que ya exista o error:", e)

# ==============================
# CREAR ARCHIVO LOCAL
# ==============================
with open(file_name, "w") as f:
    f.write("Hola AWS desde Boto3")

print("Archivo local creado")

# ==============================
# SUBIR ARCHIVO
# ==============================
try:
    s3.upload_file(file_name, bucket_name, object_name)
    print("Archivo subido correctamente")
except Exception as e:
    print("Error al subir:", e)

# ==============================
# LISTAR ARCHIVOS
# ==============================
try:
    response = s3.list_objects_v2(Bucket=bucket_name)

    print("\nArchivos en el bucket:")
    if 'Contents' in response:
        for obj in response['Contents']:
            print("-", obj['Key'])
    else:
        print("Bucket vacío")
except Exception as e:
    print("Error al listar:", e)

# ==============================
# DESCARGAR ARCHIVO
# ==============================
download_name = "descargado_" + file_name

try:
    s3.download_file(bucket_name, object_name, download_name)
    print("\nArchivo descargado como:", download_name)
except Exception as e:
    print("Error al descargar:", e)

print("\nProceso terminado")
```
